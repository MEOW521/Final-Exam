```python
#导入必要的包
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import tree
```

```python
#导入顾客信息并对分类变量进行OneHot处理
info = pd.read_excel('./顾客信息.xlsx').drop('User id', axis=1)
info = pd.get_dummies(info, prefix=['Age','Incoming','Student','Credit Rating'], columns=['Age','Incoming','Student','Credit Rating'])
```

```python
#构建决策树分类器
clf = tree.DecisionTreeClassifier(random_state=123)
X = info.iloc[:,1:]
Y = info.loc[:,'Buying']
```

```python
clf.fit(X, Y)
```

```python
#决策树分类器可视化
feature_names = info.columns[1:]
target_names = info['Buying'].unique().tolist()
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, feature_names=feature_names, class_names=target_names)
```

```python
#录入新用户信息
new = pd.DataFrame([[0,1,0,0,0,1,1,0,1,0]], columns=list(info.columns[1:]))
```

```python
#对新用户进行预测，预测结果为no：不购买
clf.predict(new)
```
